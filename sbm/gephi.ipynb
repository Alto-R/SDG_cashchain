```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# 10
colors = [
    "#A31D20",  # 1
    "#148843",  # 2
    "#4C54A0",  # 3
    "#02779B",  # 4
    "#035830",  # 5
    "#7F7CB6",  # 6
    "#6BAED5",  # 7
    "#BEBED8",  # 8
    "#9BC9DD",  # 9
    "#A2D59B",  # 10
]

# 
fig, ax = plt.subplots(figsize=(2, 4), dpi=300)

# 
legend_elements = [
    Line2D(
        [0], [0],
        marker='o',
        color='w',
        markerfacecolor=color,
        markersize=15,
        markeredgecolor='black',
        markeredgewidth=1.5,
        label=f'Community {i+1}'
    )
    for i, color in enumerate(colors)
]

# 
ax.legend(
    handles=legend_elements,
    loc='center',
    frameon=False,
    fontsize=12,
    title='Community Colors',
    title_fontsize=14,
    ncol=1,  # 
)

# 
ax.axis('off')

plt.tight_layout()
# plt.savefig('visualization/community_colors_legend.png', dpi=300, bbox_inches='tight')
plt.show()
```
