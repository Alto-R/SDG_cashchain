```python
import pandas as pd
```

```python
data = pd.read_csv('../data/202001')
data
```

```python
print(data.head(10))
print(data.columns)
```

```python
print("各列统计摘要:")
print("=" * 50)
for i, col in enumerate(data.columns, 1):
    print(f"{i}. {col}:")
    print(f"   类型: {data[col].dtype}")
    print(f"   非空值: {data[col].count()}/{len(data)}")
    print(f"   唯一值: {data[col].nunique()}")
    
    if data[col].dtype in ['int64', 'float64']:
        print(f"   范围: {data[col].min()} ~ {data[col].max()}")
        print(f"   均值: {data[col].mean():.2f}")
    print()
```

```python
# --- 2. 定义标准化的列名 ---
generic_columns = ['大分類', '中分類', '小分類']

# --- 3. 提取并重命名 発注社 的组合 ---
hatuchu_cols = ['発注社大分類', '発注社中分類', '発注社小分類']
df_hatuchu_combinations = data[hatuchu_cols]
# 重命名列，以便能和受注社数据合并
df_hatuchu_combinations.columns = generic_columns

# --- 4. 提取并重命名 受注社 的组合 ---
juchu_cols = ['受注社大分類', '受注社中分類', '受注社小分類']
df_juchu_combinations = data[juchu_cols]
# 重命名列
df_juchu_combinations.columns = generic_columns

# --- 5. 合并 (Stack) 两个 DataFrame ---
# ignore_index=True 重新创建索引
df_combined = pd.concat([df_hatuchu_combinations, df_juchu_combinations], ignore_index=True)

print(f"\n合并后的总行数 (去重前): {len(df_combined)}")

# --- 6. 找出所有唯一的组合 ---
# .drop_duplicates() 会检查所有列，只保留唯一的 (大, 中, 小) 组合行
unique_combinations = df_combined.drop_duplicates().reset_index(drop=True)

# --- 7. 显示结果 ---
print(f"找到的唯一 (大-中-小) 组合总数: {len(unique_combinations)}")

print("\n--- 唯一的行业分类组合 (前10条示例) ---")
print(unique_combinations.head(10))
unique_combinations.to_csv('unique_combinations.csv',index=None,encoding='utf-8-sig')
```
